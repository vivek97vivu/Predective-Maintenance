```python
import pandas as pd

telemetry = pd.read_csv("PdM_telemetry.csv")
failures = pd.read_csv("PdM_failures.csv")

telemetry['datetime'] = pd.to_datetime(telemetry['datetime'])
failures['datetime'] = pd.to_datetime(failures['datetime'])

telemetry['RUL'] = None

for machine in telemetry['machineID'].unique():
    machine_data = telemetry[telemetry['machineID'] == machine]
    failure_times = failures[failures['machineID'] == machine]['datetime']

    for i, row in machine_data.iterrows():
        future_failures = failure_times[failure_times > row['datetime']]
        if len(future_failures) > 0:
            next_failure = future_failures.min()
            telemetry.loc[i, 'RUL'] = (next_failure - row['datetime']).days
```

```python
from sklearn.ensemble import RandomForestRegressor

X = telemetry[['volt','rotate','pressure','vibration']]
y = telemetry['RUL']

model = RandomForestRegressor()
model.fit(X, y)
```

```python
import pandas as pd

new_data = pd.DataFrame(
    [[170,400,100,45]],
    columns=['volt','rotate','pressure','vibration']
)

model.predict(new_data)
```
